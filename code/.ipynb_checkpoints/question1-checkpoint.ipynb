```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
from statsmodels.formula.api import ols 

%matplotlib inline
```

```python
covid_df = pd.read_csv('../data/data_municipalities.csv', sep=',')

covid_df['municipality_type'] = covid_df['municipality_code']%10
covid_urban = covid_df[covid_df['municipality_type']==1]
covid_rural = covid_df[covid_df['municipality_type']==2]
covid_mix = covid_df[covid_df['municipality_type']==3]
```

```python
plt.scatter(covid_df[['population_total']], covid_df[['percent_vaccinated']], c=covid_df['municipality_type'],  cmap='viridis')
plt.xlim(0, 27000)
plt.colorbar(label='municipality_type')
```

```python
# Modeling OLS population_total vs percent_vaccinated, ignoring any distinction on municipality_type
# fit simple linear regression model 
linear_model = ols('population_total ~ percent_vaccinated', 
				data=covid_df[['percent_vaccinated','population_total']]).fit() 

# display model summary 
print(linear_model.summary()) 

# modify figure size 
fig = plt.figure(figsize=(14, 8)) 

# creating regression plots 
fig = sm.graphics.plot_regress_exog(linear_model, 
									'percent_vaccinated', 
									fig=fig) 
```

```python
# fit simple linear regression model for urban municipality_type [code 1]
linear_model_code1 = ols('population_total ~ percent_vaccinated', 
				data=covid_urban[['percent_vaccinated','population_total']]).fit() 

# display model summary 
print(linear_model_code1.summary()) 

# modify figure size 
fig = plt.figure(figsize=(14, 8)) 

# creating regression plots 
fig = sm.graphics.plot_regress_exog(linear_model_code1, 
									'percent_vaccinated', 
									fig=fig) 
```

```python
# fit simple linear regression model for rural municipality_type [code 2]
linear_model_code2 = ols('population_total ~ percent_vaccinated', 
				data=covid_rural[['percent_vaccinated','population_total']]).fit() 

# display model summary 
print(linear_model_code2.summary()) 

# modify figure size 
fig = plt.figure(figsize=(14, 8)) 

# creating regression plots 
fig = sm.graphics.plot_regress_exog(linear_model_code2, 
									'percent_vaccinated', 
									fig=fig) 
```

```python
# fit simple linear regression model for mixed municipality_type [code 3]
linear_model_code3 = ols('population_total ~ percent_vaccinated', 
				data=covid_mix[['percent_vaccinated','population_total']]).fit() 

# display model summary 
print(linear_model_code3.summary()) 

# modify figure size 
fig = plt.figure(figsize=(14, 8)) 

# creating regression plots 
fig = sm.graphics.plot_regress_exog(linear_model_code3, 
									'percent_vaccinated', 
									fig=fig) 
```

```python
# Understanding outlier
bp = covid_df[['population_total']].boxplot(rot=45, fontsize=12, figsize=(12, 8), showfliers=False, showmeans=True, meanline=True)
plt.show()
```

```python
# Calculate Q1, Q3 and IQR
Q1 = np.percentile(covid_df[['population_total']], 25)
Q3 = np.percentile(covid_df[['population_total']], 75)
IQR = Q3 - Q1

# Define the outlier thresholds
lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR

df_filtered = covid_df[(covid_df["population_total"] < upper_threshold) & (covid_df["population_total"] > lower_threshold)]
```

```python
# fit simple linear regression model with removed outlier data set
linear_model_filtered = ols('population_total ~ percent_vaccinated', 
				data=df_filtered[['percent_vaccinated','population_total']]).fit() 

# display model summary 
print(linear_model_filtered.summary()) 

# modify figure size 
fig = plt.figure(figsize=(14, 8)) 

# creating regression plots 
fig = sm.graphics.plot_regress_exog(linear_model_filtered, 
									'percent_vaccinated', 
									fig=fig) 
```
